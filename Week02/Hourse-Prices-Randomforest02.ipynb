```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('/kaggle/input/home-data-for-ml-course/train.csv')
y = train_df.SalePrice
```

```python
train_df.head(10)
```

```python
features = ['MSSubClass',
            'LotArea',
            'OverallQual',
            'OverallCond',
            'YearBuilt',
            'YearRemodAdd',
            'BsmtFinSF1',
            'BsmtFinSF2',
            'BsmtUnfSF',
            'TotalBsmtSF',
            '1stFlrSF',
            '2ndFlrSF',
            'LowQualFinSF',
            'GrLivArea',
            'BsmtFullBath',
            'BsmtHalfBath',
            'FullBath',
            'HalfBath',
            'BedroomAbvGr',
            'KitchenAbvGr',
            'TotRmsAbvGrd',
            'Fireplaces',
            'GarageCars',
            'GarageArea',
            'WoodDeckSF',
            'OpenPorchSF',
            'EnclosedPorch',
            '3SsnPorch',
            'ScreenPorch',
            'PoolArea',
            'MiscVal',
            'MoSold',
            'YrSold'
]
```

```python
X = train_df[features]
X.head(5)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state = 42)
```

# Model

```python
model_Rf = RandomForestRegressor(random_state = 42, n_estimators=700)
model_Rf.fit(X_train, y_train)
```

```python
model_pre = model_Rf.predict(X_val)
mae_val = mean_absolute_error(model_pre, y_val)
rmae_val = np.sqrt(mean_squared_error(model_pre, y_val))


print(f'mae: {mae_val}')
print(f'rmae: {rmae_val}')
```

```python
model_true = RandomForestRegressor(random_state = 42, n_estimators=700)
model_true.fit(X, y)
```

# Submission

```python
test_df = pd.read_csv('/kaggle/input/home-data-for-ml-course/test.csv')
test_df = test_df.fillna(-1)
X_test = test_df[features]
test_pre = model_true.predict(X_test)
```

```python
submit = pd.DataFrame({'Id':test_df["Id"],
                      'SalePrice':test_pre})

submit.to_csv('submission.csv', index=False)
```
